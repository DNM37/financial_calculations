```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Read the CSV file
df = pd.read_csv('financial_dataset/stock_and_mutual_fund.csv')

# Extract the closing prices
closing_prices = df['Stock Close (CAD)'].values

# Generate an array of indices representing the days
days = np.arange(len(closing_prices)).reshape(-1, 1)

# Take the natural logarithm of the closing prices
log_closing_prices = np.log(closing_prices)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(days, log_closing_prices)

# Get the line of best fit in the log scale
log_line_of_best_fit = model.predict(days)

# Transform the predictions back to the original scale
line_of_best_fit = np.exp(log_line_of_best_fit)

# Plot the closing prices and the exponential line of best fit
plt.figure(figsize=(10, 6))
plt.plot(days, closing_prices, label='Closing Prices')
plt.plot(days, line_of_best_fit, label='Exponential Line of Best Fit', color='red')
plt.xlabel('Days')
plt.ylabel('Stock Close (CAD)')
plt.title('Stock Closing Prices and Exponential Line of Best Fit')
plt.legend()
plt.show()
```
